```python
!pip install yfinance

import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load data
data = yf.download("TSLA", start="2020-01-01", end="2024-01-01")

# Keep only closing price
data = data[['Close']]

# Create prediction column (shift by 10 days)
forecast_days = 10
data['Prediction'] = data['Close'].shift(-forecast_days)

# Prepare dataset
X = np.array(data.drop(['Prediction'], axis=1))[:-forecast_days]
y = np.array(data['Prediction'])[:-forecast_days]

# Split into training and testing
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Models
lr = LinearRegression()
rf = RandomForestRegressor(n_estimators=100)

# Train
lr.fit(X_train, y_train)
rf.fit(X_train, y_train)

# Predict
lr_pred = lr.predict(X_test)
rf_pred = rf.predict(X_test)

# Evaluate
def evaluate(true, pred, name):
    mae = mean_absolute_error(true, pred)
    rmse = np.sqrt(mean_squared_error(true, pred))
    print(f"{name} MAE: {mae:.2f}")
    print(f"{name} RMSE: {rmse:.2f}\n")

evaluate(y_test, lr_pred, "Linear Regression")
evaluate(y_test, rf_pred, "Random Forest")

# Plot results
plt.figure(figsize=(10,5))
plt.plot(y_test, label="Actual")
plt.plot(lr_pred, label="Linear Regression")
plt.plot(rf_pred, label="Random Forest")
plt.legend()
plt.title("Stock Price Prediction")
plt.savefig("results.png")
plt.show()
```
